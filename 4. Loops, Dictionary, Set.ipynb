```python
cities = ["Karachi", "Quetta", "Lahore", "ISB"]
city_name = input("Enter City name: ")

for city in cities:
    if city in city_name:
        print(city_name, "is a Clean city")
```

## Nested for loop

```python
firstname = ['Iqra', 'Jayya', 'Zeenat']
lastname = ['Anwar', 'jaz', 'bro']

for first in firstname:
    for last in lastname:
        print(first +" "+last)
```

## While loop

```python
cities_to_visit = []

user_input = ''

while user_input != 'q':
    city = input("Enter city name: ")
    if city == 'q':
        user_input = 'q'
        print("Program ended!")
    else:
        cities_to_visit.append(city)
```
