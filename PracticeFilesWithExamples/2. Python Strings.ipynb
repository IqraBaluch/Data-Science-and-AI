## Indexed

```python
"{0} loves {1}. {0} works with {1}.".format("Iqra", "Python")
"{0} loves {1}. {0} works with {1}".format("Hidayat","Data")
```

## Named placeholders (cleaner)

```python
"Name: {name}, Role: {role}".format(name="Iqra", role="Analyst")
```

## Formatting numbers

```python
"Salary: {:.2f}".format(123456.789)
```


| Part | Meaning                            |
| ---- | ---------------------------------- |
| `:`  | Start of formatting rules          |
| `.2` | Number of digits **after decimal** |
| `f`  | Floating-point number              |


```python
"Counting: {:.1f}".format(123456.4565767)
```
