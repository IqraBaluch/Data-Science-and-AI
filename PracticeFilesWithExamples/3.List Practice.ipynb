# Mutability & References (THIS IS THE TRAP)

#### Step 1: The Core Truth (Read Carefully)

In Python:

- Variables do NOT store values
- Variables store references (addresses)
- Lists are mutable objects
- = does NOT copy a list

**Keep this line in your head:**

- = means ‚Äúpoint to the same object‚Äù


#### Step 2: The Classic Trap


```python
a = [1, 2, 3]
b = a

b.append(4)

print("a:", a)
print("b:", b)
```

**What happened?**

a and b point to the same list

Change via b ‚Üí reflected in a

**Visualize it:**



üëâ This is not copying. This is aliasing.

#### Step 3: PROVE IT (Very Important)

```python
print(id(a))
print(id(b))
```

Same number = same object in memory.


#### Step 4: The Safe Way (Shallow Copy)

1. Method 1: copy()

```python
a = [1, 2, 3]
b = a.copy()

b.append(4)

print(a)
print(b)
```

2. Method 2: slicing

```python
a = [1, 2, 3]
b = a[:]
```

Memory picture:


Different objects, same values.

```python
print(id(a))
print(id(b))
```

#### Step 5: THE REAL TRAP (Nested Lists)

```python
a = [[1, 2], [3, 4]]
b = a.copy()

b[0].append(99)

print("a:", a)
print("b:", b)
```

**Surprise?**

Inner lists are still shared

This is a shallow copy

**Visualization:**



#### Step 6: Deep Copy (When You REALLY Need Isolation)

```python
import copy

a = [[1, 2], [3, 4]]
b = copy.deepcopy(a)

b[0].append(99)

print("a:", a)
print("b:", b)
```

Now they‚Äôre fully independent.

#### Step 7: One-Line Mental Rules (MEMORIZE)

= ‚Üí same object

.copy() / [:] ‚Üí new list, same inner objects

deepcopy() ‚Üí fully new structure

Lists mutate in place

### üß† List vs String vs Tuple

We‚Äôll compare them by behavior, not definition.

#### 1Ô∏è‚É£ The Core Difference (One Sentence Each)
**üü¶ List** 

A mutable container for a collection of items.

**üü® String**

An immutable sequence of characters.

**üü© Tuple**

An immutable container for grouped values.

#### 2Ô∏è‚É£ Mutability (THE Big Divider)

```python
lst = [1, 2, 3]
lst[0] = 100
print(lst)
```

```python
s = "abc"
# s[0] = "x"   # ERROR
```

```python
s = "abc"
s[0] = "x"   # ERROR
```

```python
t = (1, 2, 3)
# t[0] = 100   # ERROR
```

```python
t = (1, 2, 3)
t[0] = 100   # ERROR
```

**Rule:**

Only lists can change in place

Strings & tuples cannot

#### 3Ô∏è‚É£ Memory & Safety (Why Tuples Exist)

```python
t = (1, 2, 3)
```

Tuples:

- Use less memory
- Are hashable
- Can be dictionary keys
- Are safer for fixed data
example:
coords = (24.8607, 67.0011)  # latitude, longitude


#### 4Ô∏è‚É£ Performance (Important for Pros)

- Lists ‚Üí slightly slower

- Tuples ‚Üí slightly faster

- Strings ‚Üí optimized for text operations

You won‚Äôt feel this on small data, but it matters at scale.

#### 5Ô∏è‚É£ Behavior with + (Hidden Difference)

```python
lst = [1, 2]
print(lst + [3])      # creates NEW list


s = "ab"
print(s + "c")        # creates NEW string

t = (1, 2)
print(t + (3,))       # creates NEW tuple
```

#### 7Ô∏è‚É£ When to Use WHAT (THIS IS GOLD)
